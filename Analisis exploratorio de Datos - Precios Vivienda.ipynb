```python
matrizCorrmat = df_train.corr()
f,ax = plt.subplots(figsize =(12,9))
sns.heatmap(matrizCorrmat, vmax = 1.1, square=True)
```

```python
k=10    
```

```python
cols = matrizCorrmat.nlargest(k,"SalePrice")["SalePrice"].index #toma los "K" valores mÃ¡s relacionados
cols
```

```python
cm = np.corrcoef(df_train[cols].values.T)   #transpone los valores de cols a cm
cm
```

```python
sns.set(font_scale=1.25)    #cambia el formato de las letras
```

```python
f,ax = plt.subplots(figsize =(12,9))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
```
